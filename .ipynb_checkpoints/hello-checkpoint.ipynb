```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from io import BytesIO
cameraman_img = cv2.imread(r'./image/cameraman.jpg')
plt.imshow(cameraman_img)
plt.show()
```

```python
height, width, abc = cameraman_img.shape
print(height, width, abc)

rotate_y_img = np.zeros((height, width), dtype=np.uint8)

for i in range(height):
    rotate_y_img[i, :] = cameraman_img[height - i - 1, :, 0]

plt.imshow(rotate_y_img, cmap="gray")
```
